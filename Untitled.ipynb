## 2. Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray). 

Example 1:
Input: [1,3,5,4,7]     
Output: 3 

Example 2:
Input: [2,2,2,2,2]  
Output: 1

```python
nums=[1,3,5,4,7]
def find_length_of_lcis(nums):
    if not nums:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1
            
    return max_length

print("This is the answer for the first example1: ",find_length_of_lcis(nums))
```

```python
nums1= [2,2,2,2,2]
def find_length_of_lcis(nums1):
    if not nums1:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums1)):
        if nums1[i] > nums1[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1
            
    return max_length

print("This is the answer for the second example2: ",find_length_of_lcis(nums1))
```

# 3. Given a list of non negative integers, arrange them such that they form the largest number.

 Example 1:
 Input: [10,2]
 
 Output: "210"
 
 Example 2:
 Input:[3,30,34,5,9]
 
 Output: "9534330"


```python
from typing import List

def largestNumber(nums: List[int]) -> str:
    # Convert integers to strings for comparison
    nums = [str(num) for num in nums]
    
    # Define custom sorting key to compare two numbers as strings
    nums.sort(key=lambda x: x*10, reverse=True)
    
    # Join sorted strings to form the largest number
    largest = ''.join(nums)
    
    # Remove leading zeros if any
    largest = largest.lstrip('0')
    
    # If all numbers were zeros, return '0'
    if not largest:
        return '0'
    
    return largest
```

```python
nums3 = [10,2]
print(largestNumber(nums3))
print("---------------------------------------->")
nums2 = [3, 30, 34, 5, 9]
print(largestNumber(nums2))
```

# 4. Store all the "servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
import json
import csv

# Load JSON data from file
with open("DT A1 sample_json (1) (1).json") as f:
    data = json.load(f)

# Extract "servlet-name" and "servlet-class" pairs
servlets = [(servlet['servlet-name'], servlet['servlet-class']) for servlet in data['web-app']['servlet']]

# Write to CSV file
with open('servlets.csv', 'w', newline='') as csvfile:
    fieldnames = ['servlet-name', 'servlet-class']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for servlet in servlets:
        writer.writerow({'servlet-name': servlet[0], 'servlet-class': servlet[1]})

print("CSV file generated successfully.")
```
